## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import random
import numpy as np

cnt = 0
# Simulate rolling the dice 1000 times and count the occurrences of each total
for i in range(1000):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    
    if total > 7 or total % 2 == 0:
        cnt += 1

probability = (cnt / 1000)*100
print(f"Count of totals greater than 7 or even: {cnt}")
print(f"Estimated probability: {probability}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random

# Create the box of balls: 10 white, 20 red, 30 green
ball_box = ["White"] * 10 + ["Red"] * 20 + ["Green"] * 30

#print the contents of the box and total number of balls
print("Ball box contents:", ball_box)
print(f"Total balls: {len(ball_box)}")

# Run experiment 1000 times
count_3white_2red = 0
count_all_same = 0

for experiment in range(1000):
    # Take 5 balls with replacement
    selected_balls = []
    for i in range(5):
        ball = random.choice(ball_box)
        selected_balls.append(ball)
    
    # Check if 3 white and 2 red
    if selected_balls.count("White") == 3 and selected_balls.count("Red") == 2:
        count_3white_2red += 1
    
    # Check if all same color
    if (selected_balls.count("White") == 5 or 
        selected_balls.count("Red") == 5 or 
        selected_balls.count("Green") == 5):
        count_all_same += 1

# Calculate probabilities
prob_3white_2red = (count_3white_2red / 1000) * 100
prob_all_same = (count_all_same / 1000) * 100

print(f"Probability of 3 white and 2 red: {prob_3white_2red}%")
print(f"Probability of all same color: {prob_all_same}%")
```
